```python
import requests
import pprint
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
```

```python
page = requests.get("https://coinmarketcap.com/pt-br/")
```

```python
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
print(soup.prettify())
```

```python
table = soup.find('table', attrs={'id':'currencies'})
table_body = table.find('tbody')
```

```python
print(table_body.prettify())
```

```python
rows = table_body.find_all('tr')
data = []
for row in rows:
    cols = row.find_all('td')
    cols = [ele.text.strip() for ele in cols]
    data.append([cols[1][5::],cols[2],cols[3],cols[4],cols[5],cols[6]])
```

```python
print (data)
```
